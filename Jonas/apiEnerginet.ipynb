```python
import requests
URL = 'https://api.energidataservice.dk/datastore_search?resource_id=privindustryconsumptionhour&limit=250'

with requests.Session() as S:
    DATA = dict(S.get(url=URL, headers={"user-agent": "magic browser", "Content-Type": "application/json"}).json())

print(DATA["result"]["records"])
```
